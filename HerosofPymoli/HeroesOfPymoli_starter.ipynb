```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head(5)
```

## Player Count

* Display the total number of players


```python
Total_players=len(purchase_df["SN"].unique())
pd.DataFrame({"Total Players":[Total_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_df["Item ID"].unique())
average_price=round(purchase_df["Price"].mean(),2)
Total_purchase=len(purchase_df["Purchase ID"]) 
Total_revenue=purchase_df["Price"].sum()

summary_dict={"Number of Unique Items":[unique_items],
              "Average Price":[average_price],
              "Number of Purchases":[Total_purchase],
             "Total Revenue":[Total_revenue]}
              
summary_df=pd.DataFrame(summary_dict)
summary_df["Total Revenue"]=summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df["Average Price"]=summary_df["Average Price"].map("${:,.2f}".format)
print(summary_df)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group=purchase_df.groupby("Gender")

gender_df=gender_group.agg({"SN":"nunique"}).reset_index()
gender_df["Percentage of Players"]=round((gender_df["SN"]/Total_players)*100,2)
gender_df.rename(columns={"SN":"Total Count"},inplace=True)
gender_df["Percentage of Players"]=gender_df["Percentage of Players"].map("{:,.2f}%".format)
gender_df.set_index("Gender",inplace=True)
print(gender_df)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# creating bin for ages

age_bin=[0,9,14,19,24,29,34,39,100]
bin_label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_df["Age_group"]=pd.cut(purchase_df["Age"],age_bin,labels=bin_label)

# creating agewise group for further calculation on each group

age_group=purchase_df.groupby("Age_group")

# creating a dataframe with unique names for calculation

age_df=age_group.agg({"SN": "nunique"}).reset_index()
age_df.rename(columns={"SN":"Total Count"},inplace=True)

age_df["Percentage of Players"]=round((age_df["Total Count"]/Total_players)*100,2)
age_df["Percentage of Players"]=age_df["Percentage of Players"].map("{:,.2f}%".format)
age_df.set_index("Age_group",inplace=True)
age_df.head(8)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# creating group on SN for further calculation
SN_group=purchase_df.groupby("SN")

# creating dataframes for basic calculations

SN_df1=purchase_df.groupby("SN",as_index=False).agg({"Item ID":"count","Price":"mean"})
SN_df1.rename(columns={"Item ID":"Purchase Count",
                       "Price":"Average Purchase Price"},inplace=True)

SN_df2=purchase_df.groupby("SN",as_index=False).agg({"Price":"sum"})
SN_df2.rename(columns={"Price":"Total Purchase Value",},inplace=True)

# merging and formatting above dataframes for printing summary table

SN_df=pd.merge(SN_df1,SN_df2,on="SN")
SN_df_sorted=SN_df.sort_values(by="Total Purchase Value",ascending=False)
SN_df_sorted.set_index("SN",inplace=True)
SN_df_sorted["Average Purchase Price"]=SN_df_sorted["Average Purchase Price"].map("${:,.2f}".format)
SN_df_sorted["Total Purchase Value"]=SN_df_sorted["Total Purchase Value"].map("${:,.2f}".format)
SN_df_sorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#creating dataframes for basic calculations

item_df1=purchase_df.groupby(["Item ID","Item Name"],as_index=False).agg({'Item ID': 'count', 'Price': 'sum'})
item_df1.rename(columns={"Item ID":"Purchase Count",
                         "Price":"Total Purchase Value"},inplace=True)

item_df2=purchase_df.groupby(["Item ID","Item Name"],as_index=False).agg({'Price': 'first'})

# merging, formatting and sorting above dataframes for printing summary table

item_df=pd.merge(item_df1,item_df2,on="Item Name",how="left")
sorted_df=item_df.sort_values(by="Purchase Count",ascending=False)
sorted_df.set_index("Item ID",inplace=True)

sorted_df["Price"]=sorted_df["Price"].map("${:,.2f}".format)
sorted_df["Total Purchase Value"]=sorted_df["Total Purchase Value"].map("${:,.2f}".format)
sorted_df=sorted_df[["Item Name","Purchase Count","Price","Total Purchase Value"]]

sorted_df.head()
```
